**APPLYING ELBOW TECHNIQUE**

```python
k_rng=range(1,10)
sse=[]
for k in k_rng:
  km=KMeans(n_clusters=k)
  km.fit(df[['petal_length','petal_width']])
  sse.append(km.inertia_)
```

**sse stands for sum of squared errors**

```python
sse
```

```python
plt.xlabel('k_rng')
plt.ylabel('sse')
plt.plot(k_rng,sse)
```

**APPLYING KMean ALOGRITHM**

```python
k = KMeans(n_clusters=3,random_state=0,)
y_predicted=k.fit_predict(df[['petal_length','petal_width']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head(149)
```

**ACCURACY MEASURE**

```python
from sklearn.metrics import confusion_matrix
c= confusion_matrix(df.species,df.cluster)
c
```

```python
true_labels=df.species
predicted_labels=df.cluster

c = confusion_matrix(true_labels, predicted_labels)
class_labels = ['Iris-Setosa', 'Iris-Versicolor', 'Iris-Virginica']
# plotting confusion matrix
plt.imshow(c,interpolation='nearest',cmap=plt.cm.Greens)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(len(class_labels))
plt.xticks(tick_marks,class_labels)
plt.yticks(tick_marks,class_labels)

#filling of values in matrix

for i in range(len(class_labels)):
  for j in range(len(class_labels)):
    plt.text(j,i,str(c[i][j]),ha='center',va='center',color='black')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
